```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from scipy.optimize import curve_fit

df = pd.read_csv('../../docs/datasets/execution-times.csv')

x_data = df['size']
y_data = df['time']

# Create a figure with two subplots to display both analyses side by side
def quad(x, a, b, c):
    return a * x**2 + b * x + c

# Perform curve fitting to find the optimal parameters
popt, _ = curve_fit(quad, x_data, y_data)

# Generate y values from the fitted function
fitting_curve = quad(x_data, a = popt[0], b = popt[1], c = popt[2])

plt.figure(figsize=(10, 5))
plt.scatter(x_data, y_data, label='Original Data')
plt.plot(x_data, fitting_curve, label='Fitted Curve', color='red')
plt.title('Best fitting curve for execution times')
plt.legend()
plt.xlabel('Input Size')
plt.ylabel('Time in seconds')
plt.grid(True)
plt.legend()


plt.show()
```
